```python
# PPA 2

class Triangle:

    def __init__(self, a , b , c):
        self.a=a
        self.b=b
        self.c=c

    #### Function to chek if a traingle is valid
    def Is_valid(self):
        if (self.a +self.b >self.c ) and (self.b +self.c >self.a) and (self.a +self.c >self.b):
            return "Valid"
        else:
            return "Invalid"

    #### Function to chk for the type of triangle accoding to the sides
    def Side_Classification(self):
        if self.Is_valid() == "Valid":
            if self.a == self.b == self.c :
                return "Equilateral"
            elif (self.a == self.b != self.c) or (self.a == self.c != self.b) or (self.a != self.b == self.c) :
                return "Isosceles"
            elif self.a != self.b != self.c :
                return "Scalene"

        else:
            return "Invalid"

    ### Function to chk the type of the trangle according to angles using pythagoras theorme
    def Angle_Classification(self):
        if self.Is_valid() == "Valid" :
            smallest_side , middle_side , biggest_side = sorted([self.a , self.b , self.c])

            if ( (smallest_side)**2 + (middle_side)**2 == (biggest_side)**2 ) or ( (self.a)**2 + (self.c)**2 == (biggest_side)**2 ) or ( (self.b)**2 + (self.c)**2 == (biggest_side)**2 ) :
                return "Right"

            elif ( (smallest_side)**2 + (middle_side)**2 >= (biggest_side)**2 ) or ( (smallest_side)**2 + (middle_side)**2 >= (biggest_side)**2 ) or ( (smallest_side)**2 + (middle_side)**2 >= (biggest_side)**2 ) :
                return "Acute"

            elif ( (smallest_side)**2 + (middle_side)**2 <= (biggest_side)**2 ) or ( (smallest_side)**2 + (middle_side)**2 <= (biggest_side)**2 ) or ( (smallest_side)**2 + (middle_side)**2 <= (biggest_side)**2 ) :
                return "Obtuse"

        else:
            return "Invalid"

    ### Function to get the area of the tiangle
    def Area(self):
        if self.Is_valid() == "Valid" :
            s = ( self.a + self.b + self.c ) / 2
            area = ( s * (s-self.a) * (s-self.b) * (s-self.c) ) ** 0.5
            return area

        else:
            return "Invalid"


a,b,c = 1,2,3
T=Triangle(a,b,c)
print(T.Is_valid())
print(T.Side_Classification())
print(T.Angle_Classification())
print(T.Area())
```

```python
# GRPA 1

# Write a function find_Min_Difference(L, P) that accepts a list L of integers and P (positive integer) where the size of L is greater than P. The task is to pick P different elements from the list L, where the difference between the maximum value and the minimum value in selected elements is minimum compared to other differences in possible subset of p elements. The function returns this minimum difference value.
# Note - The list can contain more than one subset of p elements that have the same minimum difference value.
# Example
# Let L = [3, 4, 1, 9, 56, 7, 9, 12, 13] and P = 5
# If we see the following two subsets of 5 elements from L
# [3, 4, 7, 9, 9] or [7, 9, 9, 12, 13]
# Here, the difference between the maximum value and the minimum value in both subset is 9 - 3 = 6 or 13 - 7 = 6 which is minimum. So the output will be 6.
# Sample Input
# [3, 4, 1, 9, 56, 7, 9, 12]
# 5
# Output
# 6


def find_Min_Difference ( L , P ):
    """
    Function to find the minimum difference between the maximum and minimun element in a subset of P element chosen form list L.

    Parameters:
    L (list): A list of intergers.
    P (int): Number of element to pick from the list.

    Returns:
    int: Minimum possible difference between the max and min element in a subset of P elemetns.

    """
    if P<=0 or P>len(L):
        return 0   # invalid case

    sorted_lsit = sorted(L)

    list_ = []
    for i in range(0 , len(L)-P+1):
        list_.append(sorted_lsit[i : i+P])

    difference_list = []
    for lst in list_ :
        difference_list.append(max(lst)-min(lst))

    return min(difference_list)




L = [3, 4, 1, 9, 56, 7, 9, 12, 13]
P = 5
print(find_Min_Difference(L,P))
```

```python
# GRPA 2


def is_prime(num):
    """
    Function to check if a number is prime.
    """
    if num < 2 :
        return False
    for i in range( 2 , (int(num**0.5)+1) ) :
        if num % i == 0 :
            return False
    return True


def Goldbach(n):
    """
    Function to  find all possible pair of prime numbers (a,b) such that:
    - Their sum is equal to the given even number n.
    - The order is maintained such that  a<=b to avoid dublicate pairs.

    Parameters:
    n (int): The even number of which we need to find prime number pairs.

    Returns:
    list: A list of tuples where each tuple (a,b) represent a valid prime number.
    """
    prime_list = []
    for i in range(2,n):
        if is_prime(i) == True :
            prime_list.append(i)

    result_ = []
    for a in prime_list:
        for b in prime_list:
            if (a+b) == n :
                result_.append(( min(a,b) , max(a,b) ))

    return list(set(result_))




n=int(input())
print(sorted(Goldbach(n)))
```

```python
# GRPA 3
def odd_one(L):
    """
    Function to determine the data type of the odd one out in a given list.

    The list contain elements where all but one are of the same data type.
    The function identifies the odd element and return its data type as a string.

    Patameters:
    L (list): A list containing at least three element of types(int, float, str or bool).

    Returns:
    str: The data type of the odd element as a string ("int" , "float" , "str" , "bool")

    """

    int_count = 0
    float_count = 0
    str_count = 0
    bool_count = 0

    for i in L:
        if type(i) == int:
            int_count += 1
        elif type(i) == str:
            str_count +=1
        elif type(i) == bool:
            bool_count += 1
        elif type(i) == float:
            float_count +=1


    if int_count == 1:
        return "int"
    elif float_count == 1:
        return "float"
    elif str_count == 1:
        return "str"
    elif bool_count == 1:
        return "bool"
    else:
        return None



print(odd_one([2,1,2,16,4.5]))
```
